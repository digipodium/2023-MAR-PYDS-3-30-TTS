Import library function

```python
from dputils.scrape import Tag, Scraper   #Tag and Scrapper are classes
```

- Tag class helps to decide what to collect from the page
- Scraper executes the scraping process 

```python
url = "https://timesofindia.indiatimes.com/news"
news_scraper = Scraper(url)
```

```python
target = Tag(cls = 'listing4')
items = Tag('li')
title = Tag('span', cls = 'w_tle')
desc = Tag('span', cls = 'w_desc')
```

```python
news_scraper.get_all(target, items, heading=title, desc=desc)
```

```python
result = news_scraper.get_all(target, items, heading=title, desc=desc)
```
